```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,y = make_moons(n_samples=1000, noise=0.1, random_state=0)

plt.figure(figsize=(12, 8))
plt.scatter(X[y == 0, 0],
            X[y == 0, 1],
            )
plt.scatter(X[y == 1, 0],
            X[y == 1, 1],
            )
plt.show()
```

```python
db = DBSCAN(eps=0.1, min_samples=3).fit(X)
y_db = db.labels_
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X[y_db == 0, 0],
            X[y_db == 0, 1],
            )
plt.scatter(X[y_db == 1, 0],
            X[y_db == 1, 1],
            )
plt.show()
```

```python
print('cluster'+ str(np.max(y_db)))
print('cluster1'+ str(len(np.where(y_db ==0)[0])))
print('cluster2'+ str(len(np.where(y_db ==1)[0])))
print('outlier'+str(len(np.where(y_db ==-1)[0])))
```

```python
from sklearn.datasets import make_blobs
NORMAL_DATA_N = 1000
args = {
        # 
        'n_samples': NORMAL_DATA_N,
        # 
        'n_features': 2,
        # 
        'centers': 3,
        # 
        'random_state': 42,
    }
X_normal, y_normal = make_blobs(**args)
```

```python
#outlier
OUTLIER_DATA_N = 100
DISTRIBUTION_RANGE = 6
X_outlier = np.random.uniform(low=-DISTRIBUTION_RANGE,
                               high=DISTRIBUTION_RANGE,
                               size=(OUTLIER_DATA_N, 2))
y_outlier = np.ones((OUTLIER_DATA_N, ))
```

```python
X_outlier += X_normal.mean(axis=0)
X = np.concatenate([X_normal, X_outlier], axis=0)
y = np.concatenate([y_normal, y_outlier], axis=0)

plt.scatter(X[y == 0, 0],
            X[y == 0, 1],
            )
plt.scatter(X[y == 1, 0],
            X[y == 1, 1],
            )
plt.scatter(X[y == 2, 0],
            X[y == 2, 1],
           )
plt.legend()
plt.show()
```

```python
db = DBSCAN(eps=1.0, min_samples=5).fit(X)
y_db = db.labels_
```

```python
print('cluster'+ str(np.max(y_db)))
print('cluster1'+ str(len(np.where(y_db ==0)[0])))
print('cluster2'+ str(len(np.where(y_db ==1)[0])))
print('cluster3'+ str(len(np.where(y_db ==2)[0])))
print('outlier'+str(len(np.where(y_db ==-1)[0])))
```

```python
plt.scatter(X[y_db == 0, 0],
            X[y_db == 0, 1],
            )
plt.scatter(X[y_db == 1, 0],
            X[y_db == 1, 1],
            )
plt.scatter(X[y_db == 2, 0],
            X[y_db == 2, 1],
           )
plt.scatter(X[y_db == 3, 0],
            X[y_db == 3, 1],
           )

plt.legend()
plt.show()
```
